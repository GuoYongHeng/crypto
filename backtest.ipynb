```python
from strategies.fix_trade_price_strategy import FixedTradeTimeStrategy
from strategies.sma_strategy import SmaStrategy
from vnpy_ctastrategy.backtesting import BacktestingEngine, OptimizationSetting
from vnpy.trader.constant import Interval
from datetime import datetime
import plotly.graph_objects as go
```

```python

engine = BacktestingEngine()

engine.set_parameters(
    vt_symbol="BTCUSDT.BINANCE",
    interval=Interval.MINUTE,
    start=datetime(2024,1,2),
    end=datetime(2024,3,20),
    rate=1/1000,
    slippage=0.02,
    size=1,
    pricetick=0.01,
    capital=30000
)

engine.add_strategy(SmaStrategy, {})
engine.load_data()
engine.run_backtesting()
engine.calculate_result()
engine.calculate_statistics()
engine.show_chart()
```

```python
setting = OptimizationSetting()
setting.set_target("sharpe_ratio")
setting.add_parameter("fast_window", 2, 20, 1)
setting.add_parameter("slow_window", 10, 40, 1)
result = engine.run_ga_optimization(setting)  # 优化策略参数
print(result)
```
